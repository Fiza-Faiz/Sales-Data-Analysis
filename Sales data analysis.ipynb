# The program analyzes the sales data of various products in different states of the United States and provides an interactive interface. Graphical representation in the form of bar graphs and pie-charts have been provided to better visualize the analysis. The programming is in python using 4 librarires. 

```python
#importing libraries and loading the data
#@author: fizaf


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from termcolor import colored,cprint

sales = pd.read_csv("A:/Business Analytics/Python/Class/BBSales.csv")

```

```python
#main funciton file 

def answers(inp):

    valid = ["1","2","3","4","5","6","7"]
    #1 Sorting by Bedding sales 
    if inp == "1":
        sort_bedding = sales.sort_values(by="Bedding", ascending=False)

        print("The first Answer  \n")
        print("A List of states sorted by Bedding sales :")
        print(sort_bedding.loc[:,['State','Bedding']])


    #2
    #baby products >> luggage
    if inp == "2":

        compare_products = sales.loc[sales.Baby > sales.Luggage]

        print("The second Answer  \n")
        print("List of states where Baby products outsold Luggage :")
        print(compare_products.loc[:,['State','Baby','Luggage']])

    # 3 curtains are popular
    if inp == "3":
        curtain_pop = sales.loc[sales.Curtains > sales["Curtains"].mean()]

        mean_curtains = sales["Curtains"].mean()
        print("The third Answer \n ")
        print("List of states where curtains are unexpectedly popular :")
        print("The average sales of curtains across states : " + str(mean_curtains))
        print(curtain_pop.loc[:,['State','Curtains']])


    #4 List of states with most popular sales category 
    if inp == "4":

        States = sales.iloc[:,0]
        #print(States)
        product_sales = sales.drop(['State','Population','PopulationPct','TotalSales'],  axis =1 )
        #print(product_sales)

        maximum = product_sales.idxmax(axis=1)
        #print(maximum)

        max_type = pd.concat([States, maximum] , axis=1)

        print("The fourth Answer \n")
        print("List of states with most popular sales category : ")
        print(max_type)




    #5 Bar Graph of top 10 Kids merchandize buying states
    if inp == "5":
        #sort_Kids = sales.sort_values(by="Bedding", ascending=False)
        sales_Kids = sales.nlargest(10, 'Kids')
        print(sales_Kids)

        #states = sales_Kids.iloc[:,0]
        y_pos = sales_Kids.State
        print(y_pos)
        values  = sales_Kids.Kids
        print(values)

        plt.bar(y_pos, values, align='center', alpha= 0.5)

        plt.xticks(rotation=45)
        plt.ylabel('Sales of Kids Goods in dollars ')
        plt.xlabel('States')
        plt.title('Top 10 States in sales of Kids Merchandize ')

        print("The fifth Answer \n ")
        print("Bar Graph of top 10 Kids merchandize buying states :")
        plt.show()





    #6 Bar Graph of sales performance in Ohio 
    
    if inp == "6" :
        sales_Ohio = sales.loc[sales.State == "Ohio"]
        Ohio_sales = sales_Ohio.drop(['Population','PopulationPct','TotalSales'],  axis =1 )
        print(Ohio_sales)


        #print(Ohio_sales.iloc[0,:])
        y_pos = list(Ohio_sales.columns.values)

        y_pos.remove('State')
        Ohio_numpy = Ohio_sales.to_numpy()
        print(Ohio_numpy)
        Ohio_numpy = Ohio_numpy[0,1:]

        print(Ohio_numpy)



        print(y_pos)

        new_x, new_y = zip(*sorted(zip(Ohio_numpy, y_pos)))

        plt.xticks(rotation=45)
        plt.bar(new_y, new_x, align='center', alpha= 0.5)

        #plt.bar(y_pos, Ohio_numpy, align='center', alpha= 0.5)

        plt.ylabel('Sales performance in dollars ')
        plt.xlabel('Category of Sales ')
        plt.title('Sales performance in Ohio ')
        plt.xticks(rotation=45)

        print("The sixth Answer \n ")
        print("Bar Graph of sales performance in Ohio : ")
        plt.show()

    #7 Sales by category in Virginia 

    if inp == "7" :
        sales_Virginia = sales.loc[sales.State == "Virginia"]
        Virginia_sales = sales_Virginia.drop(['Population','PopulationPct','TotalSales'],  axis =1 )
        print(Virginia_sales)


        #print(Virginia_sales.iloc[0,:])
        y_pos = list(Virginia_sales.columns.values)

        y_pos.remove('State')
        Virginia_numpy = Virginia_sales.to_numpy()
        print(Virginia_numpy)
        Virginia_numpy = Virginia_numpy[0,1:]

        print(Virginia_numpy)

        pie_chart = pd.DataFrame({'sales': Virginia_numpy},index = y_pos)
        plot = pie_chart.plot.pie(y='sales', figsize=(6, 6))
        plt.title('Sales by category in Virginia ')
        plt.show()



    
    if not(inp in valid ) :
        print("Invalid Input")


```

```python
#taking input from the viewer
inp = ""
while inp != "exit" :
    print(
    "Option 1 : List of states sorted by Bedding sales \n"
    "Option 2 : List of states where Baby products outsold Luggage \n"
    "Option 3 : List of states where curtains are unexpectedly popular \n"
    "Option 4 : List of states with most popular sales category \n"
    "Option 5 : Bar Graph of top 10 Kids merchandize buying states\n"
    "Option 6 : Bar Graph of sales performance in Ohio\n"
    "Option 7 : Pie chart of sales in Virginia by Category"
    )
    inp = np.array(input("Enter the option or write exit to exit : "))
    print("Option selected :" + str(inp))
    allowed = ['1','2','3','4','5','6','7','exit']

    
    if(inp.__contains__(allowed)):
        answers(str(inp))
        print(answers)
    
    
    if(not(inp.__contains__(allowed))):
        warn = colored("INVALID OPTION SELECTED !!!",attrs=['bold'])
        cprint(warn,'green','on_red')
        
    
    
```
